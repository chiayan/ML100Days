```python
import numpy as np
#將上列陣列(array1)，轉成維度為(5X6)的 array，順序按列填充。(hint:order="F")
#呈上題的 array，找出被 6 除餘 1 的數的索引
```

```python
array1 = np.array(range(30))
array1
```

```python
array1.reshape(5,6)
```

```python
array1.reshape(5,6,order="F")
######  按列(row)填充應為 order="C"; order="F"為按欄(column)填充
```

```python
array1
```

```python
array1.reshape(5,6,order="F")%6
```

```python
np.where(array1.reshape(5,6,order="F")%6==1)
```
